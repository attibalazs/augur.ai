```python
from h5py import File as h5File

from numpy import asarray

from os.path import expanduser, join

class Hdf5_esp():
    def __init__(self, file_path):
        self.field_dict = {}
        self.data_dict = {}
        
        with h5File(file_path, 'r') as self.fp:
            self.dump_h5_content()

    def dump_h5_content(self):
        for field in self.fp.keys():
            self.field_dict[field] = {}
            self.data_dict[field] = {}
            
            for esp in self.fp[field].keys():
                self.field_dict[field][esp] = list(self.fp[field][esp].keys())
                self.data_dict[field][esp] = list(self.fp[field][esp].keys())
                
                for sensor in self.fp[field][esp].keys():
                    print(self.fp[field][esp][sensor].value)
    
    def query(self, field, esp, sensor):
        return self.data_dict[field][esp][sensor]

def main():
    file_folder = join(expanduser('~'), 'data')
    file_name = 'esp_data_interp.h5'  

    file_path = join(file_folder, file_name)
    
    H5_obj = Hdf5_esp(file_path)
    
    # To print keys
    print(H5_obj.field_dict)
    
    # To query
    #print(H5_obj.query('1', '1', 'PINP').value)

if __name__ == "__main__":
    main()
```
